## FOR DEPENDENT DATA, HISTOGRAM IS PREFERED CHOICE, OTHERWISE PIE CHART IS USED TO VISUALIZER INDEPENDENT DATA

```python
# np.random.normal () - 1st para (mean), 2nd para (standard deviation), 3rd para - n different edges
```

```python
ages = np.random.normal(20,1.5,1000)
```

```python
ages
```

```python
plt.hist(ages,bins=[ages.min(),18,21,23,ages.max()],color='lightgreen')
```

```python
# Cumulitive graph
plt.hist(ages,bins=15,cumulative=True,color='lightgreen')
```

## PIE CHART

```python
langs = ['C++','C#','JAVA','PYTHON','R','JAVASCRIPT']
votes = [10,15,30,40,12,10]
```

```python
plt.pie(votes,labels=langs,autopct="%.2f%%")
plt.show()
```
