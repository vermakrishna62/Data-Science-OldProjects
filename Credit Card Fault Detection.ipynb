```python
# Under Sampling
# build a sample dataset, containing similar distribution of normal transaction and fraud transaction
```

```python
legit_sample = legit_tran.sample(n=492)
```

```python
new_credit_dataset = pd.concat([legit_sample,fraud_tran],axis='rows')
# axis 0-rows and 1-cols
```

```python
sns.countplot(data=new_credit_dataset,x='Class',width=0.2)
```

```python
new_credit_dataset.groupby('Class').mean()
```

```python
X = new_credit_dataset.drop(['Class'],axis=1).values
y = new_credit_dataset['Class'].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression().fit(X_train,y_train)
```

```python
log_reg
```

```python
log_reg.score(X_test,y_test)
```

```python
log_reg.score(X_train,y_train)
```
